````python
import os
from dotenv import load_dotenv
from smolagents import CodeAgent, LiteLLMModel, tool
import requests
import json

load_dotenv()

@tool
def get_weather(city: str) -> str:
    """L·∫•y th√¥ng tin th·ªùi ti·∫øt c·ªßa th√†nh ph·ªë
    
    Args:
        city: T√™n th√†nh ph·ªë c·∫ßn l·∫•y th√¥ng tin th·ªùi ti·∫øt
    """
    try:
        url = f"http://wttr.in/{city}?format=j1"
        response = requests.get(url)
        data = response.json()
        current = data['current_condition'][0]
        return f"Th·ªùi ti·∫øt {city}: {current['temp_C']}¬∞C, {current['weatherDesc'][0]['value']}"
    except:
        return f"Kh√¥ng th·ªÉ l·∫•y th·ªùi ti·∫øt cho {city}"

@tool
def calculate(expression: str) -> str:
    """T√≠nh to√°n bi·ªÉu th·ª©c to√°n h·ªçc
    
    Args:
        expression: Bi·ªÉu th·ª©c to√°n h·ªçc c·∫ßn t√≠nh to√°n
    """
    try:
        result = eval(expression)
        return f"{expression} = {result}"
    except:
        return "Bi·ªÉu th·ª©c kh√¥ng h·ª£p l·ªá"

from duckduckgo_search import DDGS
from typing import List

@tool
def search_web(query: str) -> str:
    """
    T√¨m ki·∫øm th√¥ng tin tr√™n web b·∫±ng DuckDuckGo (ddgs)

    Args:
        query: T·ª´ kh√≥a c·∫ßn t√¨m ki·∫øm
    """
    try:
        results: List[str] = []
        with DDGS() as ddgs:
            for r in ddgs.text(query, region="wt-wt", safesearch="moderate", max_results=3):
                results.append(f"- {r['title']}\n  {r['href']}\n  {r['body']}")
        if results:
            return f"K·∫øt qu·∫£ cho '{query}':\n\n" + "\n\n".join(results)
        else:
            return f"Kh√¥ng t√¨m th·∫•y k·∫øt qu·∫£ n√†o cho '{query}'."
    except Exception as e:
        return f"L·ªói khi t√¨m ki·∫øm: {e}"

model = LiteLLMModel(model_id="gemini/gemini-1.5-flash", api_key=os.getenv("GEMINI_API_KEY"))

agent = CodeAgent(
    tools=[get_weather, calculate, search_web], 
    model=model,
    max_steps=5
)

def chat():
    print("=" * 50)
    print("ü§ñ AI AGENT ƒê∆Ø·ª¢C K√çCH HO·∫†T!")
    print("=" * 50)
    print("T√¥i c√≥ th·ªÉ:")
    print("‚Ä¢ Xem th·ªùi ti·∫øt: 'Th·ªùi ti·∫øt H√† N·ªôi?'")
    print("‚Ä¢ T√≠nh to√°n: 'T√≠nh 123 * 456'")
    print("‚Ä¢ T√¨m ki·∫øm: 'T√¨m hi·ªÉu v·ªÅ Python'")
    print("‚Ä¢ G√µ 'exit' ƒë·ªÉ d·ª´ng")
    print("-" * 50)
    
    while True:
        try:
            user_input = input("\nüí¨ B·∫°n h·ªèi: ").strip()
            
            if not user_input:
                print("‚ùå Vui l√≤ng nh·∫≠p c√¢u h·ªèi!")
                continue
                
            if user_input.lower() in ['exit', 'quit', 'tho√°t', 'bye']:
                print("üëã T·∫°m bi·ªát!")
                break
            
            print("üîÑ Agent ƒëang x·ª≠ l√Ω...")
            result = agent.run(user_input)
            print(f"ü§ñ Agent: {result}")
            
        except KeyboardInterrupt:
            print("\nüëã T·∫°m bi·ªát!")
            break
        except Exception as e:
            print(f"‚ùå C√≥ l·ªói x·∫£y ra: {e}")

if __name__ == "__main__":
    chat()
````

````python
from flask import Flask, request, jsonify
from smolagent import CodeAgent
from smolagent.tools import ToolRegistry, tool
from smolagent.llms import OpenAI

# 1. ƒê·ªãnh nghƒ©a tool
@tool
def get_weather(city: str) -> str:
    return f"{city}: Tr·ªùi n·∫Øng 32¬∞C"  # Gi·∫£ l·∫≠p tr·∫£ k·∫øt qu·∫£

# 2. T·∫°o Flask app
app = Flask(__name__)

# 3. Kh·ªüi t·∫°o agent
tools = ToolRegistry(tools=[get_weather])
llm = OpenAI()  # t·ª± ƒë·ªông l·∫•y t·ª´ OPENAI_API_KEY env var

agent = CodeAgent(
    name="WeatherAssistant",
    role="Assistant tr·∫£ l·ªùi th√¥ng tin th·ªùi ti·∫øt.",
    tools=tools,
    llm=llm
)

# 4. API endpoint
@app.route('/ask', methods=['POST'])
def ask():
    user_input = request.json.get("question")
    result = agent.chat(user_input)
    return jsonify({"response": result})

if __name__ == '__main__':
    app.run(port=5005)
````
