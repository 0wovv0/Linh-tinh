````python
import os
from dotenv import load_dotenv
from smolagents import CodeAgent, LiteLLMModel, tool
import requests
import json

load_dotenv()

@tool
def get_weather(city: str) -> str:
    """Láº¥y thÃ´ng tin thá»i tiáº¿t cá»§a thÃ nh phá»‘
    
    Args:
        city: TÃªn thÃ nh phá»‘ cáº§n láº¥y thÃ´ng tin thá»i tiáº¿t
    """
    try:
        url = f"http://wttr.in/{city}?format=j1"
        response = requests.get(url)
        data = response.json()
        current = data['current_condition'][0]
        return f"Thá»i tiáº¿t {city}: {current['temp_C']}Â°C, {current['weatherDesc'][0]['value']}"
    except:
        return f"KhÃ´ng thá»ƒ láº¥y thá»i tiáº¿t cho {city}"

@tool
def calculate(expression: str) -> str:
    """TÃ­nh toÃ¡n biá»ƒu thá»©c toÃ¡n há»c
    
    Args:
        expression: Biá»ƒu thá»©c toÃ¡n há»c cáº§n tÃ­nh toÃ¡n
    """
    try:
        result = eval(expression)
        return f"{expression} = {result}"
    except:
        return "Biá»ƒu thá»©c khÃ´ng há»£p lá»‡"

from duckduckgo_search import DDGS
from typing import List

@tool
def search_web(query: str) -> str:
    """
    TÃ¬m kiáº¿m thÃ´ng tin trÃªn web báº±ng DuckDuckGo (ddgs)

    Args:
        query: Tá»« khÃ³a cáº§n tÃ¬m kiáº¿m
    """
    try:
        results: List[str] = []
        with DDGS() as ddgs:
            for r in ddgs.text(query, region="wt-wt", safesearch="moderate", max_results=3):
                results.append(f"- {r['title']}\n  {r['href']}\n  {r['body']}")
        if results:
            return f"Káº¿t quáº£ cho '{query}':\n\n" + "\n\n".join(results)
        else:
            return f"KhÃ´ng tÃ¬m tháº¥y káº¿t quáº£ nÃ o cho '{query}'."
    except Exception as e:
        return f"Lá»—i khi tÃ¬m kiáº¿m: {e}"

model = LiteLLMModel(model_id="gemini/gemini-1.5-flash", api_key=os.getenv("GEMINI_API_KEY"))

agent = CodeAgent(
    tools=[get_weather, calculate, search_web], 
    model=model,
    max_steps=5
)

def chat():
    print("=" * 50)
    print("ğŸ¤– AI AGENT ÄÆ¯á»¢C KÃCH HOáº T!")
    print("=" * 50)
    print("TÃ´i cÃ³ thá»ƒ:")
    print("â€¢ Xem thá»i tiáº¿t: 'Thá»i tiáº¿t HÃ  Ná»™i?'")
    print("â€¢ TÃ­nh toÃ¡n: 'TÃ­nh 123 * 456'")
    print("â€¢ TÃ¬m kiáº¿m: 'TÃ¬m hiá»ƒu vá» Python'")
    print("â€¢ GÃµ 'exit' Ä‘á»ƒ dá»«ng")
    print("-" * 50)
    
    while True:
        try:
            user_input = input("\nğŸ’¬ Báº¡n há»i: ").strip()
            
            if not user_input:
                print("âŒ Vui lÃ²ng nháº­p cÃ¢u há»i!")
                continue
                
            if user_input.lower() in ['exit', 'quit', 'thoÃ¡t', 'bye']:
                print("ğŸ‘‹ Táº¡m biá»‡t!")
                break
            
            print("ğŸ”„ Agent Ä‘ang xá»­ lÃ½...")
            result = agent.run(user_input)
            print(f"ğŸ¤– Agent: {result}")
            
        except KeyboardInterrupt:
            print("\nğŸ‘‹ Táº¡m biá»‡t!")
            break
        except Exception as e:
            print(f"âŒ CÃ³ lá»—i xáº£y ra: {e}")

if __name__ == "__main__":
    chat()
````

````python
from flask import Flask, request, jsonify
from smolagent import CodeAgent
from smolagent.tools import ToolRegistry, tool
from smolagent.llms import OpenAI

# 1. Äá»‹nh nghÄ©a tool
@tool
def get_weather(city: str) -> str:
    return f"{city}: Trá»i náº¯ng 32Â°C"  # Giáº£ láº­p tráº£ káº¿t quáº£

# 2. Táº¡o Flask app
app = Flask(__name__)

# 3. Khá»Ÿi táº¡o agent
tools = ToolRegistry(tools=[get_weather])
llm = OpenAI()  # tá»± Ä‘á»™ng láº¥y tá»« OPENAI_API_KEY env var

agent = CodeAgent(
    name="WeatherAssistant",
    role="Assistant tráº£ lá»i thÃ´ng tin thá»i tiáº¿t.",
    tools=tools,
    llm=llm
)

# 4. API endpoint
@app.route('/ask', methods=['POST'])
def ask():
    user_input = request.json.get("question")
    result = agent.chat(user_input)
    return jsonify({"response": result})

if __name__ == '__main__':
    app.run(port=5005)
````
